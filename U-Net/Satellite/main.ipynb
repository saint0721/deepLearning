```python
import numpy as np
import json, os 
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
from matplotlib.patches import Rectangle
from skimage.io import imread
```

```python
map_base_dir = './input/'
map_img_dir = './input/train/images/'
```

```python
json_path = os.path.join(map_base_dir, 'annotation.json')
with open(json_path, 'r') as f:
    annot_data = json.load(f)
```

```python
image_df = pd.DataFrame(annot_data['images'])
image_df.sample(3)
fig, m_axes = plt.subplots(2, 2, figsize = (10, 10))

for c_ax, (_, c_row) in zip(m_axes.flatten(), image_df.sample(4).iterrows()):
    img_data = imread(os.path.join(map_img_dir, c_row['file_name']))
    c_ax.imshow(img_data)
```
